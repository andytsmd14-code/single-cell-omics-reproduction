## Part 1 — Human Cell Atlas (HCA)

The HCA dataset comprises 10,125 samples across 15 tissue categories. Nearly 70% of samples lack ancestry annotation; among the 3,043 samples with reported ancestry, European ancestry is overrepresented roughly 6-fold relative to the global population reference, while Asian, African, and Latino ancestries are all underrepresented.

**Scripts:** `01_clean_hca.py` → `02_analysis_hca.py` → `03_tissue_stack_bar_hca.py`


### Figure 1B — Ancestry vs. global reference

![Figure 1B](figures/figure1B_reproduction.png)

Representation ratios (observed ÷ expected): European ~6.4×, Asian 0.38×, African 0.63×, Latino 0.48× — closely matching the paper's reported ~6× European overrepresentation. Chi-square goodness-of-fit test: p ≈ 0 (paper reports p < 2e-16).


### Figure 1C — Ancestry by tissue type

![Figure 1C](figures/figure1C_reproduction.png)

European ancestry dominates almost every tissue category; Immune and Gut show comparatively higher Asian representation, consistent with the paper's text (Immune 41.7% Asian, Gut 41.6% Asian in the original analysis).


### Figure 1D — Sex composition vs. global reference

![Figure 1D](figures/figure1D_reproduction.png)

chi2 = 7.09, p = 0.0078 — a statistically significant, though modest, deviation from the 50/50 global sex reference.


### Figure 1E — Sex composition by tissue type

![Figure 1E](figures/figure1E_reproduction.png)

Breast and reproductive tissues show the expected strong female skew; Eye and Musculoskeletal tissues are notably male-enriched, matching the paper's description.

## Part 2 — Human Tumor Atlas Network (HTAN) vs. SEER

The HTAN dataset comprises 1,959 tumor samples across cancer types. Among the 1,719 samples with reported ancestry, European ancestry is 78.3% of the cohort — reproduced here using the actual SEER*Explorer 2018-2022 reference data (not an approximation), matching the paper's methodology exactly.

**Scripts:** `04_clean_htan.py`, `08_clean_seer.py` → `09_htan_vs_seer.py`


### Figure 2 — HTAN vs. SEER, ancestry and sex composition

![Figure 2](figures/figure2_reproduction.png)

**A/B (ancestry):** every cancer type's chi-square test against the SEER ancestry reference is highly significant (p « 0.001), confirming pervasive European overrepresentation across all cancer types.

**C/D (sex):** most cancer types show no significant sex-composition deviation from SEER, **except Liver** (p = 0.00045) — matching the paper's specific mention of a female skew in liver tumors (68.4% female in HTAN).

## Part 3 — Reanalysis: HTAN vs. Taiwan's National Cancer Registry

**This is the reanalysis component of the project.** The original paper compares HTAN against the US SEER reference. Here, the reference population is substituted with Taiwan's national cancer incidence data (data.gov.tw, dataset 6399), to test whether the sex-representation findings are specific to the US population context.

**Scripts:** `05_taiwan_reference.py` → `07_htan_vs_taiwan.py`

**Important limitation:** Taiwan's cancer registry does not use the US-style ancestry/race categorization framework, so this reanalysis is limited to sex composition only — the ancestry comparison could not be extended to the Taiwan context. This is a genuine methodological limitation of cross-national reproduction, not an error in either dataset.


### HTAN vs. Taiwan — sex composition by cancer type

![HTAN vs Taiwan](figures/figure_htan_vs_taiwan_sex.png)

**Key finding:** switching the reference population from SEER (US) to Taiwan's registry changes *which* cancer types show statistically significant sex skew.

- **Liver cancer** shows a much larger sex disparity against the Taiwan reference (male-dominant, reflecting Taiwan's higher hepatitis B/C-related liver cancer burden in men) than against SEER.
- **Colorectal cancer** becomes statistically significant against the Taiwan reference (p = 0.011), which was not specifically flagged in the original paper's US-referenced analysis.

This illustrates a broader point: conclusions about "representation gaps" are reference-population-dependent, not absolute properties of the dataset alone.

## Part 4 — PsychAD vs. ADRD Reference

The PsychAD dataset comprises 1,494 samples across 9 neuropsychiatric/neurodegenerative diagnostic categories. Ancestry and sex composition were reproduced essentially exactly from the raw metadata (European 65.9%, African 22.6%, Latino 9.0%, Asian 1.8%, female 51.6% — all matching the paper's reported figures).

**Scripts:** `10_clean_psychad.py`, `11_clean_adrd_reference.py` → `12_psychad_figures.py`


### Figure 3 — PsychAD ancestry and sex composition

![Figure 3](figures/figure3_reproduction.png)

**Sex chi-square tests** (PsychAD vs. literature-derived disease-specific sex ratios) show **Dementia with Lewy Bodies** (p = 1.22e-6) and **Vascular Dementia** (p = 2.01e-7) as the most significantly sex-skewed diagnoses, consistent with the paper's emphasis on these two conditions.

**A documented reproduction gap (Panel B):** our computed ADRD ancestry composition from the raw Zenodo-hosted case counts (count ÷ total) yields 81.3% European, vs. 59.1% reported in the paper's text. Cross-checking the original source (Matthews et al., 2019) shows that publication reports ADRD **prevalence rates within each ancestry group** (e.g., 11.3% of white Medicare beneficiaries have ADRD) — a different statistic from ancestry **composition** of the ADRD-affected population. Converting between the two requires 2014 US population-by-race data as an additional weighting step, which was not obtainable in time (the U.S. Census Bureau API requires a personal API key). This is left as an open, documented limitation rather than resolved — see the README for the full account of this investigation.
